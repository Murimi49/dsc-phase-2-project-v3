Merging Tables

```python
#1. Merge directors and persons table on person id
merged_directors_df= pd.merge(persons_df, directors_df, on="person_id", how='inner')
```

```python
merged_directors_df
```

```python
#2. Merge movie_basics with movie_ratings on movie_id
merged_movie_info_df= pd.merge(movie_basics_df, movie_ratings_df, on="movie_id", how='inner')
```

```python
final_movie_table_df= pd.merge(merged_directors_df, merged_movie_info_df ,on="movie_id", how='inner')
final_movie_table_df
```

```python
#Save Data and visualize graphs with external tool(tableau)
final_movie_table_df.to_csv('clean_final_movie_table_df', index=False)
```
